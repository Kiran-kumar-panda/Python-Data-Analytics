```python
# average danceability percentage
df['danceability_%'].mean()
```

Average percent of danceabilty of those tracks is approx 67

```python
mean1=df[df['released_year']==2019]['danceability_%'].mean()
mean2=df[df['released_year']==2020]['danceability_%'].mean()
mean3=df[df['released_year']==2021]['danceability_%'].mean()
mean4=df[df['released_year']==2022]['danceability_%'].mean()
mean5=df[df['released_year']==2023]['danceability_%'].mean()
```

```python
l1=[mean1,mean2,mean3,mean4,mean5]
l2=[2019,2020,2021,2022,2023]
data=pd.DataFrame({"Year":l2,"Danceability_%":l1})
plt.figure(figsize=(10,6))
plt.xlabel('Year')
plt.ylabel('Danceability %')
plt.title("Last 5 years vs danceability %")

ax=sns.barplot(x=l2,y=l1,data=data)
```

# This is the percentage of danceability of last 5 years

Hence, we conclude from here that danceabilty percentage is increasing every year
